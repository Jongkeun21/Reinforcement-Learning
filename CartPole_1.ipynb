## CartPole_Q-Network

    --> state만 제공하고, 가능한 action을 모두 가져오는 방법

```python
import tensorflow as tf
import numpy as np
import gym

tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
env = gym.make('CartPole-v0')
```

```python
x = tf.placeholder(shape=[1, 4], dtype=tf.float32)
y = tf.placeholder(shape=[1, env.action_space.n], dtype=tf.float32)
lr = tf.placeholder(dtype=tf.float32)

inputs = env.observation_space.shape[0]
```

```python
xavier_ = tf.contrib.layers.xavier_initializer()

W1 = tf.get_variable(shape=[inputs, 3], initializer=xavier_, name='W1')
W2 = tf.get_variable(shape=[3, 9], initializer=xavier_, name='W2')
W3 = tf.get_variable(shape=[9, 3], initializer=xavier_, name='W3')
W4 = tf.get_variable(shape=[3, env.action_space.n], initializer=xavier_, name='W4')

L1 = tf.nn.relu(tf.matmul(x, W1))
L2 = tf.nn.relu(tf.matmul(L1, W2))
L3 = tf.nn.relu(tf.matmul(L2, W3))
Q_ = tf.matmul(L3, W4)
```

```python
initial_rate = 1e-1
num_episodes = 300
e = 0.1
df = 0.99
rList = []
```

```python
loss = tf.reduce_sum(tf.square(y-Q_))
optimizer = tf.train.AdamOptimizer(learning_rate=lr)
train = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()
decay = []

with tf.Session() as sess :
    sess.run(init)
    
    for i in range(num_episodes) :
        s = env.reset()
        e = 1./((i/50) + 10)
        rAll = 0
        j = 0
        d = False
        s_t = sess.run(tf.expand_dims(s, axis=0))
        
        while not d :
            j += 1
            
            Q = sess.run(Q_, feed_dict={x: s_t, lr: initial_rate})
            
            if e > np.random.rand(1) :
                a = env.action_space.sample()
            else :
                a = np.argmax(Q)
                
            s1, r, d, _ = env.step(a)
            
            if d :
                Q[0,a] = -100
            else :
                s1_t = sess.run(tf.expand_dims(s1, axis=0))
                Q1 = sess.run(Q_, feed_dict={x: s1_t, lr: initial_rate})
                Q[0,a] = r + (df*np.argmax(Q1))
                
            _ = sess.run(train, feed_dict={x: s_t, y: Q, lr: initial_rate})
            rAll += r
            s_t = s1_t
            
        rList.append(rAll)
#         decay.append(rAll)
        
#         if len(decay) == 10 :
#             if sum(decay) >= rAll*10 :
#                 initial_rate *= 0.1
#                 print("============================ changed ============================")
                
#             decay = []
            
        print("Episode {} finished after {} timesteps with Reward={}. Score: {}".format(i, j, rAll, np.mean(rList)))
```

```python
print("Score: ", sum(rList)/num_episodes)
```
